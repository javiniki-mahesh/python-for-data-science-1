# Plot formatting

```python
from pandas import DataFrame, read_csv

import matplotlib.pyplot as plt
from matplotlib.pyplot import rcParams
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 5,5
```

```python
x = range(1,10)
y = [2,4,6,8,2,8,6,4,2]

plt.bar(x,y)
```

```python
color = 'green'
width = [.6,.6,.6,.8,.8,.8,.6,.6,.6]
plt.bar(x,y,width = width, color = color,align='center')

# width is to define width
# color is to define color
# align is to make alignment
```

```python
plt.plot(y)
```

```python
plt.plot(y,ls = 'dotted')
```

```python
#all the line styles
plt.subplot(1,4,1)
plt.plot(y,ls=':')

plt.subplot(1,4,2)
plt.plot(y,ls='-')

plt.subplot(1,4,3)
plt.plot(y,ls='--')

plt.subplot(1,4,4)
plt.plot(y,ls='-.')
```

```python
#all the line styles
plt.subplot(1,4,1)
plt.plot(y,ls='dotted')

plt.subplot(1,4,2)
plt.plot(y,ls='solid')

plt.subplot(1,4,3)
plt.plot(y,ls='dashed')

plt.subplot(1,4,4)
plt.plot(y,ls='dashdot')
```

```python
#different marker styles
plt.subplot(1,4,1)
plt.plot(y,ls='dotted',marker = '^',mew = 10)

plt.subplot(1,4,2)
plt.plot(y,ls='solid',marker = '*',mew = 10)

plt.subplot(1,4,3)
plt.plot(y,ls='dashed',marker = 'D',mew = 10)

plt.subplot(1,4,4)
plt.plot(y,ls='dashdot',marker = 'o',mew = 10)

# marker edge width is more to identify quickly
```

```python
plt.subplot(1,4,1)
plt.plot(y,linewidth= 3,marker = 'o',markersize = 10,color='green')

plt.subplot(1,4,2)
plt.plot(y,linewidth= 3,marker = 'v',markersize = 10,color='red')

plt.subplot(1,4,3)
plt.plot(y,linewidth= 3,marker = '4',markersize = 10,color='blue')

plt.subplot(1,4,4)
plt.plot(y,linewidth= 3,marker = '*',markersize = 10,color='yellow')
```

```python
plt.plot(y,linewidth= 5,marker = 'o',markersize = 20,color='green')
```

# Defining elements of a plot

```python
import matplotlib.pyplot as plt

from matplotlib.pyplot import rcParams
```

```python
%matplotlib inline 
# to fit plot in this jupyter notebook 
rcParams['figure.figsize'] = 5,5
# fixed dimensions of the plot
```

# Defining axes, ticks, and grids

```python
x = range(1,10)
y = [2, 4, 6, 8, 0, 8, 6, 4, 2]

fig = plt.figure()

ax = fig.add_axes([.1,.1,1,1])
ax.plot(x,y)
```

```python
x = range(1,10)
y = [2, 4, 6, 8, 0, 8, 6, 4, 2]

fig = plt.figure()

ax = fig.add_axes([.5,.5,3,1]) # change in here
ax.grid()
ax.plot(x,y)
```

```python
x = range(1,10)
y = [2, 4, 6, 8, 0, 8, 6, 4, 2]

fig = plt.figure()

ax = fig.add_axes([.5,.5,1,3]) # change in here
ax.grid()
ax.plot(x,y)
```

```python
x = range(1,10)
y = [2, 4, 6, 8, 0, 8, 6, 4, 2]

fig = plt.figure()

ax = fig.add_axes([2,.5,1,1]) # change in here
ax.grid()
ax.plot(x,y)
```

```python
x = range(1,10)
y = [2, 4, 6, 8, 0, 8, 6, 4, 2]

fig = plt.figure()

ax = fig.add_axes([2,2,1,1]) # change in here
ax.grid()
ax.plot(x,y)
```

```python
fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])

ax.set_xlim([1,10])
ax.set_ylim([0,10])

ax.set_xticks([0,1,2,4,5,6,8,9])
ax.set_yticks([0,1,2,3,4,5,8,10])
# manually setting specific ticks

ax.grid()
ax.plot(x,y)
```

```python
fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])

ax.set_xlim([1,10])
ax.set_ylim([0,10])

# ax.set_xticks([0,1,2,4,5,6,8,9])
# ax.set_yticks([0,1,2,3,4,5,8,10])
ax.set_xticks(range(0,11))
ax.set_yticks(range(0,10))
# setting ticks in range

ax.grid()
ax.plot(x,y)
```

# Genereting multiple plots in one figure with subplots

## General way of subplotting

```python
plt.subplot(1,2,1)
```

```python
plt.subplot(1,2,1)
# here 1,2 is represents devide 1 figure into two parts
# 1 represents, in those 2 parts 1 fig is this one
plt.plot(x)
```

```python
plt.subplot(1,2,1)
plt.plot(x)
plt.plot(y)
```

```python
plt.subplot(1,2,1)
plt.plot(x)

plt.subplot(1,2,2)
plt.plot(y)
```

```python
fig = plt.figure()
fig, (ax1, ax2) = plt.subplots(1,2)
print('x',x)
ax1.plot(x)
ax2.plot(y)
plt.show()
```

```python
fig = plt.figure()
fig, (ax1, ax2) = plt.subplots(1,2)
print('x',x)
ax1.plot(x)
ax2.plot(y)
ax1.plot(y)
plt.show()
```

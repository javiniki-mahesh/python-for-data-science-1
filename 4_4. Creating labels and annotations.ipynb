# Creating labels and annotations

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import rcParams

from pandas import DataFrame, read_csv
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 10,4
```

# Labeling plot features

### The Functional method

```python
x = ['apple', 'samsung','realMe']
y = range(1,4)

plt.bar(x,y)
# labels for x and y
plt.xlabel("Mobiles",fontsize = 20)
plt.ylabel("prices",fontsize = 20)

plt.show()
```

```python
print(x,y,sep='\n')
```

```python
rcParams['font.size'] = 15
#labels size is fixed 15
# because I felt the names are not visible
```

```python
plt.pie(y,labels=x) # set labels using argument labels
plt.title("Tech companies")
# fixed a title for pie chart
plt.show()
```

```python
address = "D:\data wrangling linked in learning\Data\mtcars.csv"
cars = read_csv(address)
l = ['car_names']
#print(df.columns)
#cars.columns = ['car_names','mpg','cyl','disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']
# above are column names in the data set 
# logic to get above from table
for i in range(1,len(cars.columns)):
    l.append(cars.columns[i])
print(l)
print(cars.columns)
# using .columns also we could get columns of data set but to add a single column name
# which is to replace 0 to car_names we have done this process
```

```python
cars
```

```python
cars.columns = l
cars.columns
```

```python
print(type(cars))
print(type(cars.mpg))
# cars is a data frame so we could access the columns using . operator
# when we use cars.mpg it gives mpg column which acts as series
mpg = cars.mpg
mpg
```

### The object-oriented method

```python
# object oriented method of plotting
# which could be start with balnk plot 
# then after we keep adding depending on the requirement
fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])

mpg.plot()
```

```python
rcParams['font.size'] = 10

fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])

mpg.plot()

ax.set_xticks(range(32))
ax.set_xticklabels(cars.car_names, rotation = 60, fontsize = 'medium')
ax.set_title('Miles per Gallon of Cars in mtcars Dataset', fontsize = 20)

ax.set_xlabel('Car names', fontsize = 15)
ax.set_ylabel('Miles/Gal', fontsize = 15)
```

## Adding a legend to plot

### The object-oriented method

```python
fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])

mpg.plot()

ax.set_xticks(range(32))
ax.set_xticklabels(cars.car_names, rotation = 60, fontsize = 'medium')
ax.set_title('Miles per Gallon of Cars in mtcars Dataset')

ax.set_xlabel('car names')
ax.set_ylabel('miles/gal')

ax.legend(loc = 'best')
```

# Annotating plot

```python
mpg.max()
```

```python
fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])

mpg.plot()

ax.set_xticks(range(32))
ax.set_xticklabels(cars.car_names, rotation = 60, fontsize = 'medium')
ax.set_title('Miles per Gallon of Cars in mtcars Dataset')

ax.set_xlabel('car names')
ax.set_ylabel('miles/gal')

ax.legend(loc = 'best')
ax.set_ylim([0,45])

ax.annotate('Toyota', xy = (19,33.9), xytext = (21,36),
           arrowprops = dict(facecolor = 'black', shrink = 0.05))
# adding text in the plot where we need to give more information to audience
```

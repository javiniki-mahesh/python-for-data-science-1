```python
from pandas import read_csv
```

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import rcParams
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 8,8
```

## The simplest time  series plot

```python
address = "D:\data wrangling linked in learning\Data\Superstore-Sales.csv"
# In this data set Order Dates are encoded by cp1252 by microsoft Excel 
# or by anycase dataframe is encoded as cp1252
df = read_csv(address, index_col = 'Order Date',encoding = 'cp1252', parse_dates = True)
# dates are parsed by parameter parse_dates
df
```

```python
df['Order Quantity'].plot()
```

```python
df2 = df.sample(n = 100, random_state = 25, axis = 0)
#print(df2.count())
plt.xlabel("Order date")
plt.ylabel("Order quantity")
plt.title("Superstore sales")
df2['Order Quantity'].plot()
```

# Creating statistical data graphics

```python
from pandas import read_csv
```

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import rcParams
```

```python
import seaborn as sb
# seaborn library
sb.set_style('darkgrid')
# setting the style to darkgrid
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 6,6
```

## Data distribution with histograms

```python
address = "D:\data wrangling linked in learning\Data\mtcars.csv"

cars = read_csv(address)

cars.columns = ['car_names','mpg','cyl','disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']

cars.index = cars.car_names

mpg = cars['mpg']

mpg.plot(kind = 'hist')
```

```python
plt.hist(mpg)
plt.plot()
```

```python
sb.distplot(mpg)
```

```python
sb.histplot(mpg)
```

```python
sb.displot(mpg)
```

### Scatter plot

```python
cars.plot(kind = 'scatter', x = 'hp', y = 'mpg', c = ['darkgray'], s = 150)
```

```python
sb.regplot(x = 'hp', y = 'mpg', data = cars, scatter = True)
```

## Scatter plot matrix

```python
sb.pairplot(cars)
```

```python
cars_subset = cars[['mpg', 'disp', 'hp', 'wt']]
sb.pairplot(cars_subset)
plt.show()
```

### Boxplot

```python
cars.boxplot(column = 'mpg', by = 'am')
cars.boxplot(column = 'wt', by = 'am')
```

```python
sb.boxplot(x = 'am', y = 'mpg', data = cars, palette = 'hls')
```

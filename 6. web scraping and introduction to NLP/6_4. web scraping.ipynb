```python
import urllib.request
from bs4 import BeautifulSoup
import re
```

```python
response = urllib.request.urlopen('https://analytics.usa.gov/').read()
print(response)
```

```python
print(type(response))
# above starting b refers the text is in binary format
```

```python
html = BeautifulSoup(response,'lxml')
print(type(html))
```

```python
print(html.prettify())
```

```python
print(html.text.strip())
```

```python
for i in range(len(html.text.strip().split('\n'))):
    if html.text.strip().split('\n')[i] != '':
        print(html.text.strip().split('\n')[i])
# when we use split it form a list of elements with delimiter of \n
# from that we are getting the output in a better readable manner
```

```python
tags = html.find_all()
for tag in tags:
    print(tag.name,end = '\t')
```

```python
html.find_all('a')
# a tags where all links present
```

```python
http_links = html.find_all('a',attrs={'href':re.compile('^http')})
# links starts with http
http_links
```

```python
for http in http_links:
    print(http)
print("\n\n\n",type(http))
print(type(http_links))
```

```python
link = html.find_all('a')
for l in link:
    print(l['href'])
```

```python
with open('links from website.txt','w') as file:
    for l in link:
        print(l['href'])
        file.write(l['href'])
        file.write('\n')
```
